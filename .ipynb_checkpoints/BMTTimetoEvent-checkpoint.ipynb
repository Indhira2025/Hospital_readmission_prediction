```python
import numpy as np
import pandas as pd
from scipy.stats import linregress



# 2. Engineered Wearable Metrics (Calculated on sorted dataframe)
df_15d['step_volatility_7d'] = (
    df_15d.groupby('STUDY_PRTCPT_ID')['total_steps']
    .transform(lambda x: x.rolling(window=7, min_periods=3).std())
    / 
    (df_15d.groupby('STUDY_PRTCPT_ID')['total_steps']
    .transform(lambda x: x.rolling(window=7, min_periods=3).mean()) + 1e-5)
)

df_15d['sleep_efficiency'] = (df_15d['ASLEEP_VALUE'] / (df_15d['INBED_VALUE'] + 1)).clip(0, 1)
df_15d['circadian_blunting_index'] = (df_15d['mean_hr'] - df_15d['night_hr']) / (df_15d['mean_hr'] + 1)

# 3. Robust Trend Slope Feature Function
def calculate_slope(series):
    # Remove NaNs safely first
    clean_series = series.dropna()
    if len(clean_series) < 5: 
        return 0.0
    # Match the x-axis length perfectly to the clean y-axis length
    slope, _, _, _, _ = linregress(range(len(clean_series)), clean_series)
    return slope

# Extract step trend per patient
df_trends = df_15d.groupby('STUDY_PRTCPT_ID')['total_steps'].apply(calculate_slope).reset_index(name='step_trend_slope')

# 4. Standard Aggregation
agg_features = {
    'total_steps': ['mean', 'sum'],
    'percent_active': ['mean'],
    'moderately_active': ['mean'],
    'morning_steps': ['mean'],
    'mean_steps_per_minute': ['mean'],
    'night_hr': ['mean', 'std'],  
    'mean_hr': ['mean'],
    'circadian_blunting_index': ['mean', 'min'],
    'step_volatility_7d': ['max'],
    'sleep_efficiency': ['mean', 'min', 'std'],
    'ASLEEP_VALUE': ['max'],
    'MOOD': ['mean', 'std'],
    'DaysFromTransplant': ['max'],
}

df_agg = df_15d.groupby('STUDY_PRTCPT_ID').agg(agg_features)
df_agg.columns = ['_'.join(col).strip() for col in df_agg.columns]
df_agg = df_agg.reset_index()

# 5. Merge the Trend Slopes into Aggregated Features Matrix
df_agg = df_agg.merge(df_trends, on='STUDY_PRTCPT_ID', how='left')

# 6. Diagnosing the 27 Missing Patients (Left Join Verification)
df_diagnostic = df_targets_final[['STUDY_PRTCPT_ID', 'readmission_label', 'time_to_event']].merge(
    df_agg, on='STUDY_PRTCPT_ID', how='left'
)

missing_patients = df_diagnostic[df_diagnostic['total_steps_mean'].isna()]['STUDY_PRTCPT_ID'].tolist()
print(f"Total Patients in Targets: {len(df_targets_final)}")
print(f"Patients with valid 15-day wearables: {len(df_agg)}")
print(f"Number of Patients completely missing wearable rows: {len(missing_patients)}")

# 7. Finalize Model Cohort (Dropping the wearable-missing rows)
df_model_ready1 = df_diagnostic.dropna(subset=['total_steps_mean']).copy()

print(f"\nFinal Patients ready for survival model: {len(df_model_ready1)}")
print(f"Final Class distribution:\n{df_model_ready1['readmission_label'].value_counts()}")
```

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

from sksurv.linear_model import CoxnetSurvivalAnalysis
from sksurv.metrics import concordance_index_censored


# ================================
# 1. FEATURES
# ================================
selected_features = [
    'total_steps_mean',
    #'total_steps_sum',
    'percent_active_mean',
    #'moderately_active_mean',
    'morning_steps_mean',
    #'mean_steps_per_minute_mean',
    #'step_trend_slope',
    'step_volatility_7d_max',

    #'mean_hr_mean',
    #'night_hr_mean',
    'night_hr_std',
    'circadian_blunting_index_mean',
    'circadian_blunting_index_min',

    #'sleep_efficiency_mean',
    'sleep_efficiency_min',
    #'sleep_efficiency_std',
    #'ASLEEP_VALUE_max'
]

X = df_model_ready1[selected_features].copy()

# Survival target (structured format)
y = np.array(
    list(zip(
        df_model_ready1['readmission_label'].astype(bool),
        df_model_ready1['time_to_event']
    )),
    dtype=[('event', '?'), ('time', '<f8')]
)


# ================================
# 2. CROSS-VALIDATION SETUP
# ================================
kf = KFold(n_splits=5, shuffle=True, random_state=42)

c_indices = []
all_coefficients = []



best_alpha = None
best_train_c = -np.inf

for alpha_idx, alpha in enumerate(coxnet.alphas_):
    risk_train = X_train_scaled @ coxnet.coef_[:, alpha_idx]
    c_train = concordance_index_censored(
        y_train['event'], y_train['time'], risk_train
    )[0]
    if c_train > best_train_c:
        best_train_c = c_train
        best_alpha_idx = alpha_idx



print(" Running Properly Tuned Penalized Cox (Coxnet) Survival Model...")

# ================================
# 3. MODEL LOOP
# ================================
for fold, (train_idx, test_idx) in enumerate(kf.split(X)):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    # Imputation
    imputer = SimpleImputer(strategy='median')
    X_train_imp = imputer.fit_transform(X_train)
    X_test_imp  = imputer.transform(X_test)

    # Scaling
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_imp)
    X_test_scaled  = scaler.transform(X_test_imp)

    # Fit Coxnet
    coxnet = CoxnetSurvivalAnalysis(
        l1_ratio=0.5, alpha_min_ratio=0.01,
        n_alphas=100, max_iter=5000
    )
    coxnet.fit(X_train_scaled, y_train)

    # ✅ Select best alpha using TRAIN set only (inside fold)
    best_train_c   = -np.inf
    best_alpha_idx = 0

    for alpha_idx in range(len(coxnet.alphas_)):
        risk_train = X_train_scaled @ coxnet.coef_[:, alpha_idx]
        c_train = concordance_index_censored(
            y_train['event'], y_train['time'], risk_train
        )[0]
        if c_train > best_train_c:
            best_train_c   = c_train
            best_alpha_idx = alpha_idx

    # ✅ Evaluate on TEST set using the chosen alpha
    risk_test   = X_test_scaled @ coxnet.coef_[:, best_alpha_idx]
    best_c_index = concordance_index_censored(
        y_test['event'], y_test['time'], risk_test
    )[0]
    best_coef = coxnet.coef_[:, best_alpha_idx]

    c_indices.append(best_c_index)
    all_coefficients.append(best_coef)

# ================================
# 5. RESULTS
# ================================
mean_c_index = np.mean(c_indices)
std_c_index = np.std(c_indices)

print("\n📊 ================= RESULTS ================= 📊")
print(f"Mean C-index: {mean_c_index:.4f}")
print(f"Std C-index : {std_c_index:.4f}")
print(f"95% CI      : [{mean_c_index - 1.96*std_c_index:.4f}, {mean_c_index + 1.96*std_c_index:.4f}]")
print("=================================================\n")


# ================================
# 6. COEFFICIENT SUMMARY
# ================================
mean_coefs = np.mean(np.vstack(all_coefficients), axis=0)

df_coefficients = pd.DataFrame({
    "Feature": selected_features,
    "Mean_Coefficient": mean_coefs,
    "Abs_Impact": np.abs(mean_coefs)
}).sort_values("Abs_Impact", ascending=False)

print("🧬 ================= IMPORTANT FEATURES ================= 🧬")
print(df_coefficients)
```

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

from sksurv.ensemble import RandomSurvivalForest
from sksurv.metrics import concordance_index_censored
from sklearn.inspection import permutation_importance

# ================================
# 1. FEATURES (SAME AS  COX MODEL)
# ================================
selected_features = [
    'total_steps_mean',
    #'total_steps_sum',
    'percent_active_mean',
    'morning_steps_mean',
    #'step_trend_slope',
    'step_volatility_7d_max',
    'night_hr_std',
    'circadian_blunting_index_mean',
    'circadian_blunting_index_min',
    'sleep_efficiency_min',
    #'MOOD_std'
    #'sleep_efficiency_std'
]

X = df_model_ready1[selected_features].copy()

# Survival target
y = np.array(
    list(zip(
        df_model_ready1['readmission_label'].astype(bool),
        df_model_ready1['time_to_event']
    )),
    dtype=[('event', '?'), ('time', '<f8')]
)


# ================================
# 2. CROSS-VALIDATION
# ================================
kf = KFold(n_splits=5, shuffle=True, random_state=42)

c_indices = []
feature_importances = []

print(" Running Random Survival Forest Pipeline...")


# ================================
# 3. MODEL LOOP
# ================================
for fold, (train_idx, test_idx) in enumerate(kf.split(X)):

    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    # -------- Missing value imputation --------
    imputer = SimpleImputer(strategy='median')

    X_train_imp = imputer.fit_transform(X_train)
    X_test_imp = imputer.transform(X_test)

    # ================================
    # 4. RANDOM SURVIVAL FOREST MODEL
    # ================================
    rsf = RandomSurvivalForest(
        n_estimators=300,
        min_samples_split=10,
        min_samples_leaf=5,
        max_features="sqrt",
        n_jobs=-1,
        random_state=42 + fold
    )

    rsf.fit(X_train_imp, y_train)

    # ================================
    # 5. PREDICTION + C-INDEX
    # ================================
    risk_scores = rsf.predict(X_test_imp)

    c_index = concordance_index_censored(
        y_test['event'],
        y_test['time'],
        risk_scores
    )[0]

    c_indices.append(c_index)

    print(f"Fold {fold+1} C-index: {c_index:.4f}")

    # ================================
    # 6. PERMUTATION FEATURE IMPORTANCE
    # ================================
    perm_result = permutation_importance(
        rsf,
        X_test_imp,
        y_test,
        n_repeats=10,
        random_state=42
    )

    feature_importances.append(
        perm_result.importances_mean
    )

# ================================
# 6. RESULTS
# ================================
mean_c_index = np.mean(c_indices)
std_c_index = np.std(c_indices)

print("\n📊 ================= RESULTS ================= 📊")
print(f"Mean C-index: {mean_c_index:.4f}")
print(f"Std C-index : {std_c_index:.4f}")
print(f"95% CI      : [{mean_c_index - 1.96*std_c_index:.4f}, {mean_c_index + 1.96*std_c_index:.4f}]")
print("=================================================\n")


# ================================
# 7. FEATURE IMPORTANCE
# ================================
mean_importance = np.mean(np.vstack(feature_importances), axis=0)

df_importance = pd.DataFrame({
    "Feature": selected_features,
    "Importance": mean_importance
}).sort_values("Importance", ascending=False)

print("🌲 ================= IMPORTANT FEATURES (RSF) ================= 🌲")
print(df_importance)
```

sleep_efficiency_min: Captures worst recovery state: In BMT patients, clinical deterioration often first manifests as “poor nights” characterized by fragmented and low-efficiency sleep; the minimum Sleep Efficiency Index captures this most disrupted physiological state, which may serve as an early warning signal prior to readmission.

circadian_blunting_index_min: Captures worst circadian disruption state: The minimum circadian blunting index is computed from
Circadian Blunting Index=(mean_hr−night_hr)/mean_hr+1
and then taking the minimum value across the monitoring period, which identifies the day-night period with the smallest physiologic separation (i.e., greatest circadian disruption).Thismeasures how strongly heart rate drops at night compared to daytime. So it reflects circadian (day–night) autonomic separation.


step_volatility_7d_max : The highest level of day-to-day variation in a patient’s step count over any 7-day period, reflecting how inconsistent their physical activity was during recovery .
The maximum 7-day step volatility is derived from
Step Volatility7d = SD(7-day steps)/( Mean(7-day steps)+10−5)
and then taking the maximum value across the monitoring period, which identifies the period of greatest fluctuation in daily physical activity. 
This reflects the most unstable behavioral recovery phase, where patients show highly inconsistent mobility patterns that may indicate physiologic stress, fatigue cycles, or intermittent clinical decline.


| #      | Feature                                  | Computation                                      | Clinical meaning                                                | Key literature evidence                                                                |
| ------ | ---------------------------------------- | ------------------------------------------------ | --------------------------------------------------------------- | -------------------------------------------------------------------------------------- |
| **1**  | Circadian rhythm amplitude (activity/HR) | `amplitude = (max_24h - min_24h)` or cosinor fit | Strong circadian rhythm = better recovery; flattening = frailty | Wearable circadian disruption linked to disease risk and outcomes ([ScienceDirect][1]) |
| **2**  | Circadian blunting index                 | `1 - normalized amplitude`                       | Loss of rhythm organization → physiological stress              | Cancer + HCT rhythm disruption studies ([ScienceDirect][1])                            |
| **3**  | Step trend slope (7–14 day)              | `LinearRegression(time, steps).coef_`            | Functional recovery trajectory (key predictor of readmission)   | Wearable oncology recovery studies ([MDPI][2])                                         |
